```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
a = []
with open('./historyhistory_lr0.1.pkl', 'rb') as f:
    a = pickle.load(f)
print(len(a['val_accuracy']))
print(a.keys())
```

```python
x = range(1,len(a['val_accuracy'])+1)
```

```python
# fig = plt.figure()
fig = plt.figure(figsize=(15,10))
plt.xlabel('Epochs', fontsize=16)
plt.ylabel('Accuracy', fontsize=18)
y = a['val_accuracy']
for i in range(len(a['val_accuracy'])):
    if i % 10 == 0:
        plt.annotate(s = '%0.2f'%y[i], xy = (x[i], y[i]), fontsize = 8)

plt.plot(x,y, label = 'On Test Data')
y = a['accuracy']
for i in range(len(a['val_accuracy'])):
    if i % 10 == 0:
        plt.annotate(s = '%0.2f'%y[i], xy = (x[i], y[i]), fontsize = 8)
plt.plot(x,y, label = 'On Train Data')

plt.legend(loc=2, fontsize="small")

plt.savefig('./acc_lrtest.png',bbox_inches = 'tight')
```
